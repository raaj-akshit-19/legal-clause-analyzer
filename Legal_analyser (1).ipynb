```python
import re
import json
from google.colab import files

def analyze_clauses(text):
    clauses = []

    for match in re.finditer(r'\b(?:if|in case)\b (.*?)(?:,| then | may | shall | will )(.*?)(?:\.|;|$)',
                             text, re.IGNORECASE):
        condition, consequence = match.groups()
        clauses.append({
            "type": "Condition Clause",
            "condition": condition.strip(),
            "consequence": consequence.strip()
        })

    for match in re.finditer(r'\bunless\b (.*?)(?:,| may | shall | will )(.*?)(?:\.|;|$)',
                             text, re.IGNORECASE):
        condition, consequence = match.groups()
        clauses.append({
            "type": "Exception Clause",
            "condition": condition.strip(),
            "consequence": consequence.strip()
        })

    for match in re.finditer(r'\bprovided that\b (.*?)(?:,| may | shall | will )(.*?)(?:\.|;|$)',
                             text, re.IGNORECASE):
        condition, consequence = match.groups()
        clauses.append({
            "type": "Conditional Clause",
            "condition": condition.strip(),
            "consequence": consequence.strip()
        })

    return clauses
uploaded = files.upload()

file_name = next(iter(uploaded))

with open(file_name, "r") as f:
    text = f.read()

clauses = analyze_clauses(text)
for clause in clauses:
    print(f"[{clause['type']}]")
    print(f"Condition: {clause['condition']}")
    print(f"Consequence: {clause['consequence']}\n")

print("JSON Output:")
print(json.dumps(clauses, indent=2))
```
